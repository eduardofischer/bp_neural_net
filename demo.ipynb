```python
# Funções auxiliares
def print_list_of_lists(list):
    for e in list:
        print(['{:.3f}'.format(n) for n in e])
```

```python
import numpy as np
import neural_net as nn

inputs = np.random.uniform(low=0.0, high=1.0, size=5)

net = nn.NeuralNetwork([1, 2, 1])
net.plot()
net.train([[.7]], [[1]])
```

```python
print('Network: \n{}\n'.format(net.network))
```

```python
print('Weights:')
print(net.weights)
```

```python
print('Deltas:')
print_list_of_lists(net.deltas)
```

```python
print('Gradients:')
print(net.gradients)
```

```python
print('Outputs:')
print_list_of_lists(net.outputs)
```
